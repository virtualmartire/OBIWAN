In this notebook we compare the MD throughput performance of the models considered in our work. MACE-OFF23 data are extracted from 10.48550/arxiv.2312.15211.

```python
import matplotlib.pyplot as plt
```

```python
x_axis = [99, 198, 498, 999, 1998]

obiwan = [6.11, 5.34, 3.77, 2.48, 1.27]
anakin = [5.99, 5.23, 4.01, 2.70, 1.51]
mace_off23 = [None, 6.6, 4.4, 2.6, 1.4]
```

```python
# set the size of all the fonts to 14
plt.rc('font', size=14)

plt.plot(x_axis, obiwan, label='obiwan', color='blue')
plt.scatter(x_axis, obiwan, color='blue')

plt.plot(x_axis, anakin, label='anakin', color='red')
plt.scatter(x_axis, anakin, color='red')

plt.plot(x_axis, mace_off23, label='mace_off23', color='purple', linestyle='dashed')
plt.scatter(x_axis, mace_off23, color='purple')

plt.xlabel('Number of atoms')
plt.ylabel('ns/day')
plt.title('MD throughput comparison')
plt.legend()

plt.grid(color='0.9')
plt.gca().set_axisbelow(True)

fig = plt.gcf()
fig.set_size_inches(10, 8)
```
